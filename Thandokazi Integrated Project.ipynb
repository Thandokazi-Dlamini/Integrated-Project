```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv("insurance_claims.csv")
```

```python
df.head()
```

```python
null_counts = df.isnull().sum()
null_counts
```

```python
df_new = df.drop('_c39', axis=1, errors='ignore')
```

```python
df_new.head()
```

```python
df_new.describe()
```

```python
# Choose the column for the histogram
column_name = 'age'

# Plot the histogram
plt.hist(df[column_name], bins=3, edgecolor='black')

# Add labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

# Display the histogram
plt.show()
```

```python
bin_edges = [0, 30, 55, 100]  # Define the bin edges
bin_labels = ['Young Adult', 'Middle Aged', 'Elderly']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['ages_category'] = pd.cut(df_new['age'], bins=bin_edges, labels=bin_labels)
```

```python
df_new.head()
```

```python
df_new["Contract Years"] = df_new["months_as_customer"]/12
```

```python
df_new.head()
```

```python
df_new['net_value_of_customer'] = df_new['policy_annual_premium'] - df_new['total_claim_amount']
```

```python
df_new.head()
```

```python
df_new['incident_date'] = pd.to_datetime(df_new['incident_date'])
df_new['season'] = df_new['incident_date'].dt.month % 12 // 3 + 1
```

```python
df_new.head()
```

```python
current_year = pd.to_datetime('today').year
df_new['policyholder_experience'] = current_year - df_new['auto_year']
```

```python
df_new.head()
```

```python
df_new['vehicle_age'] = current_year - df_new['auto_year']
```

```python
df_new.head()
```

```python
df_new['claim_ratio'] = df_new['total_claim_amount'] / df_new['policy_annual_premium']
```

```python
df_new.head()
```

```python
df_new['incident_month'] = df_new['incident_date'].dt.month
```

```python
df_new.head()
```

```python
plt.scatter(df_new['age'], df_new['policy_annual_premium'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Policy Annual Premium')
plt.title('Age vs. Policy Annual Premium')
plt.show()
```

```python
claim_type_counts = df_new['incident_type'].value_counts()
plt.pie(claim_type_counts, labels=claim_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Claim Types')
plt.show()
```

```python
total_loss_threshold = 0.8  # Adjust the threshold as needed
df_new['total_loss'] = (df_new['total_claim_amount'] > total_loss_threshold * df_new['vehicle_claim']).astype(int)
```

```python
df_new.head()
```

```python
fraud_rate_by_age = df_new.groupby('age_group')['fraud_reported'].mean()
fraud_rate_by_age.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Age Group')
plt.ylabel('Fraud Rate')
plt.title('Fraud Rate by Age Group')
plt.show()
```

```python
import seaborn as sns

claim_amounts_corr = df_new[['total_claim_amount', 'injury_claim', 'property_claim', 'vehicle_claim']].corr()
sns.heatmap(claim_amounts_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Claim Amounts Correlation Heatmap')
plt.show()


```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='auto_make', data=df_new, palette='Set3', order=df_new['auto_make'].value_counts().index)
plt.xlabel('Auto Make')
plt.ylabel('Count')
plt.title('Distribution of Vehicles by Make')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='ages_category', y='total_claim_amount', data=df_new, hue='fraud_reported', palette='coolwarm')
plt.title('Age Category vs. Total Claim Amount')
plt.xlabel('Age Category')
plt.ylabel('Total Claim Amount')
plt.show()
```

```python
current_year = pd.to_datetime('today').year
df_new['vehicle_age'] = current_year - df_new['auto_year']
```

```python
df_new.head()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='insured_education_level', y='policy_annual_premium', data=df_new, palette='muted')
plt.xlabel('Education Level')
plt.ylabel('Policy Annual Premium')
plt.title('Distribution of Annual Premiums by Education Level')
plt.show()
```

```python
df_new['multiple_vehicles_involved'] = df_new['number_of_vehicles_involved'] > 1
```

```python
df_new.head()
```

```python
age_bins = [18, 30, 40, 50, 60, 100]
age_labels = ['18-30', '31-40', '41-50', '51-60', '61+']
df_new['policy_holder_age_group'] = pd.cut(df_new['age'], bins=age_bins, labels=age_labels)
```

```python
df_new.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='police_report_available', y='total_claim_amount', data=df_new, palette='mako')
plt.xlabel('Police Report Availability')
plt.ylabel('Average Claim Amount')
plt.title('Average Claim Amount by Police Report Availability')
plt.show()
```

```python
avg_premium_by_education = df_new.groupby('insured_education_level')['policy_annual_premium'].mean().reset_index()
avg_premium_by_education.columns = ['insured_education_level', 'avg_annual_premium_by_education']
df_new = pd.merge(df_new, avg_premium_by_education, on='insured_education_level', how='left')

```

```python
df_new.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='insured_sex', y='policy_annual_premium', data=df_new, palette='viridis')
plt.xlabel('Gender')
plt.ylabel('Average Annual Premium')
plt.title('Average Annual Premium by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='incident_severity', data=df_new, palette='pastel')
plt.xlabel('Incident Severity')
plt.ylabel('Count')
plt.title('Incident Severity Count')
plt.show()
```
